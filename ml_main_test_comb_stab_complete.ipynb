# THIS CODE CREATES ALL POSSIBLE COMBINATION OF CATEGORICAL VALUES IN THE DATASET FOR PREDICTION PURPOSE

```python
import pandas as pd
import numpy as np
```

```python
from itertools import combinations
```

```python
from itertools import product
```

```python
main_df=pd.read_csv("cert_perov_main_w_stability.csv")
```

```python
# main_list = list(product(*main))
```

```python
main_df.columns
```

```python
useless_cols = ['Paper #','Device #','DOI','Manufacturer',
       'Manuscripted submitted year', 'Country','Passive_layer','Certified year']
num_cols = ['Voc (V)','Voc per cell (V)', 'Isc (mA)', 'Jsc (mA/cm^2)', 'FF (%)', 'Pmax (mW)',
       'Vmax (V)', 'Imax (mA)', 'Apeture area (cm^2)','extra',
       'extra%', 'MA', 'FA', 'Pb', 'I', 'Br','exposed_time', 'retained_per']
cat_cols = ['Device size type',
        'Transparent conductive oxide type',
       'Electron transport layer type', 'Perovskite composition',
       'Perovskite chemistry type', 'Perovksite depostion method',
       'Hole transport layer type', 'Conductive electrode layer',
       'Device structure']
tar_cols = ['Efficiency (%)','stability']
```

```python
use_df = main_df[num_cols]
```

```python
use_df[cat_cols] = main_df[cat_cols]
use_df[tar_cols] = main_df[tar_cols]
```

```python
use_df.shape
```

```python
use_df.to_csv('usable_df.csv',index=False)
```

# MAKING THE COMBINATION DATASET

```python
Device_size_type_u = use_df['Device size type'].unique()
TCO_type_u = use_df['Transparent conductive oxide type'].unique()
ETL_type_u = use_df['Electron transport layer type'].unique()
perov_comp = use_df['Perovskite composition'].unique()
Perov_chem_type_u = use_df['Perovskite chemistry type'].unique()
Perov_depo_type_u = use_df['Perovksite depostion method'].unique()
HTL_type_u = use_df['Hole transport layer type'].unique()
conductive_elctrode_type_u = use_df['Conductive electrode layer'].unique()
Device_struct_type_u = use_df['Device structure'].unique()
```

```python
scem_list = [list(Device_size_type_u),list(TCO_type_u),list(ETL_type_u),list(perov_comp),list(Perov_chem_type_u),list(Perov_depo_type_u),list(HTL_type_u),list(conductive_elctrode_type_u),list(Device_struct_type_u)]
```

```python
scremble = list(product(*scem_list))
```

```python
len(scremble)
```

```python
screm_data = pd.DataFrame(scremble,columns=cat_cols)
```

```python
 screm_data.to_csv('screm_text.csv',index=False)
```

# LABEL ENCODING

```python
cat_df = use_df[cat_cols]
```

```python
from sklearn.preprocessing import LabelEncoder
import joblib
```

```python
# cat_df['Device size type'].unique(),screm_data['Device size type'].unique()
```

```python
for col in cat_df.columns:
    le = LabelEncoder()
    le.fit(cat_df[col])
    joblib.dump(le,f"./serializer_object/label_encoder_categorical/{col}_encoder.pkl")
    cat_df[col] = le.transform(cat_df[col])
    
```

```python
for col in screm_data.columns:
    le_trans = joblib.load(f"./serializer_object/label_encoder_categorical/{col}_encoder.pkl")
    screm_data[col] = le_trans.transform(screm_data[col])
```
